### Minimal

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
from ipywidgets import interact

@interact(a=(-4,4,0.01))
def plot_integral(a=-1):
    x = np.linspace(-4, 4, 100)
    _, axs = plt.subplots(1,2, figsize=(8, 4))
    axi = axs.flat
    
    plt.sca(next(axi))
    plt.plot(x, norm.pdf(x))
    interval = x[x<=a]
    plt.fill_between(interval, norm.pdf(interval), color='red')
    plt.title("""Probability Density Function (f)""")
    plt.ylabel("$f(x)$")
    plt.xlabel("$x$")
    p = norm.cdf(a)
    plt.text(4, 0.4, "$\int_{-\infty}^a f(x)dx=$" + f"{p:.2f}", fontsize=10, color="red", va="top", ha="right")
    
    plt.sca(next(axi))
    plt.plot(x, norm.cdf(x))
    plt.plot([-4, a], [p, p], 'r', ls=':')
    plt.plot([a, a], [0, p], 'r', ls=':')
    plt.text(-4, p, f'$F(a) = {p:.2f}$', va='bottom', c='r')
    plt.scatter(a, p, c='r')
    plt.title("Cumulative Density Function (F)")
    
    plt.ylabel("$F(x)$")
    plt.xlabel("$x$")
    
    plt.suptitle("$P(X≤a)=$" f"{p:.2f}", y=0, va="top", c='r', size='14')
    plt.tight_layout()
```

### With titles

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
from ipywidgets import interact

def draw_point(fun, x):
    plt.vlines(x, 0, fun(x), colors='k', linestyles=":", linewidths=2)
    plt.hlines(fun(x), -4, x, colors='k', linestyles=":", linewidths=2)
    plt.plot(x, fun(x), "o", c='r')
    
    plt.text(x+ 0.1, 0, f"${x:.2f}$")
    plt.text(-4, fun(x)+0.01, f"${fun(x):.2f}$")

@interact
def plot_integral(a=(-4,4,0.01)):
    x = np.linspace(-4, 4, 100)
    _, axs = plt.subplots(1,2, figsize=(12, 6))
    axi = axs.flat
    
    plt.sca(next(axi))
    plt.plot(x, norm.pdf(x))
    interval = x[x<=a]
    plt.fill_between(interval, norm.pdf(interval), color='red')
    plt.title("""Probability Density Function""")
    plt.ylabel("$f(x)$")
    plt.xlabel("$x$")
    plt.grid()
    
    plt.sca(next(axi))
    plt.plot(x, norm.cdf(x))
    draw_point(norm.cdf, a)
    plt.title("Cumulative Density Function")
    plt.ylabel("$F(x)$")
    plt.xlabel("$x$")
    plt.grid()
    
    plt.tight_layout()
```

### Integral features

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
from ipywidgets import interact

@interact(a=(-4,4,0.01), b=(-4,4,0.01))
def plot_integral(a=-1, b=1):
    #plt.figure(figsize=(8, 4))
    # plot the curve
    x = np.linspace(-4, 4, 100)
    plt.plot(x, norm.pdf(x))
    # plot the area
    interval = x[(x>=a) & (x<=b)]
    plt.fill_between(interval, norm.pdf(interval), color='red')
    # annotate the graph
    plt.ylabel("$f(x)$")
    plt.xlabel("$x$")
    plt.text(-4, 0.3, f"$\int_a^b f(x)dx={norm.cdf(b) - norm.cdf(a):.3f}$", fontsize=14)
```
