```python
from scipy.stats import norm
from scipy.stats import t
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
```

```python
X = np.linspace(-4, 4, 200)
```

```python
@interact 
def compare_norm_t(df = np.arange(1,101)):
    X = np.linspace(-4, 4, 200)
    plt.plot(X, norm.pdf(X), X, t.pdf(X, df));
    plt.title(f'''Différence relative maximale
entre N(0,1) et T{df}(0,1):
{
              (((norm.pdf(X) - t.pdf(X, df))/norm.pdf(X)).max()) * 100
              :.2f} %''')
compare_norm_t()
```

## Probability distribution for a `continuous` random variable

```python
x = np.linspace(-5, 5, 100)
```

```python
y = stats.norm.pdf(x)
```

```python
plt.plot(x, y)
plt.scatter
plt.xlabel("x")
plt.ylabel("probability density");
```

**Read this as: "a random value of x (drawn from this distribution) has a probability of being between between 1 and 2 of: *area under the curve between 1 and 2*"**

Note that the total area under the curve is equal to 1

```python
y = stats.norm.cdf(x)
```

```python
plt.plot(x, y)
plt.scatter
plt.xlabel("x")
plt.ylabel("Probability");
```

**Read this as: "a random value of x (drawn from this distribution) has a probability of being inferior to -1 is equal: *the value of cdf at -1*"**
