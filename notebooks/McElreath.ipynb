## Berkson's paradox
6.1


```python
N = 200
p = 0.1
nw = stats.norm().rvs(N)
tw = stats.norm().rvs(N)
s = nw + tw
q = np.quantile(s, 1-p)
selected = (s>q)

sns.scatterplot(x=tw[~selected], y=nw[~selected], label="not selected")
sns.regplot(x=tw[selected], y=nw[selected], color="red", label="selected")
plt.xlabel("trustworthiness")
plt.ylabel("newsworthiness")
plt.legend()
```
